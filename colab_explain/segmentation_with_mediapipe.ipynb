```python
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt

segmentation = mp.solutions.selfie_segmentation
```

```python
img=cv2.imread("test.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
with segmentation.SelfieSegmentation(model_selection=0) as filter:
    results = filter.process(img)

plt.imshow(results.segmentation_mask)
```

```python
mask=results.segmentation_mask
back=cv2.imread("back.png")
back=cv2.resize(back,(img.shape[1],img.shape[0]))
back=cv2.cvtColor(back,cv2.COLOR_BGR2RGB)

test_img=img.copy()
test_img[np.where(mask<0.2)]=back[np.where(mask<0.2)]
plt.imshow(test_img)
```

```python
test1_img=img.copy()
test1_back=cv2.GaussianBlur(img,(0,0),15)

test1_img[np.where(mask<0.2)]=test1_back[np.where(mask<0.2)]
plt.imshow(test1_img)
```

```python
uni=cv2.imread("universe.png")
uni=cv2.resize(uni,(img.shape[1],img.shape[0]))
uni=cv2.cvtColor(uni,cv2.COLOR_BGR2RGB)

test2_img=img.copy()
test2_img[np.where(mask>0.2)]=uni[np.where(mask>0.2)]
plt.imshow(test2_img)
```

```python
test3_img=img.copy()
test3_back=cv2.GaussianBlur(img,(0,0),15)

test3_img[np.where(mask>0.2)]=test1_back[np.where(mask>0.2)]
plt.imshow(test3_img)
```
